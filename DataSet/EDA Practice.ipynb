# Titanic Exploratory Data Analysis

## Phase 1: Environment Setup & Data Loading
*In this phase, I import the necessary libraries and load the raw Titanic dataset to understand its initial structure.*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('dataset/titanic/train.csv')
data = pd.DataFrame(data)
data.head(n=10)
```

```python
np.shape(data)
```

```python
data.columns
```

```python
data.info()
```

## Phase 2: Data Cleaning & Type Conversion
*This phase ensures that our categorical data is formatted correctly for mathematical analysis, such as correlation calculations.*

```python
sns.heatmap(data.isnull(), yticklabels=False)
# sns.set_palette()
# np.shape(missing_count)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Age'] = data['Age'].astype(int)
data.head(10)
```

```python
data.drop(columns=['Cabin'], inplace=True)
data
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data.head()
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
if data.duplicated().sum() > 0:
    print(data.duplicated().sum())
else:
    print('No duplicate found')
```

```python
data['Sex'] = data['Sex'].astype('category') 
data['Pclass'] = data['Pclass'].astype('category') 
data['Embarked'] = data['Embarked'].astype('category')
data['Survived'] = data['Survived'].astype('category')
```

```python
data.info()
```

```python
col_names = ['Age', 'Fare', 'SibSp','Parch']
category_col = data[col_names]
category_col = category_col.astype('int')
data[col_names] = category_col
data.info()
```

## Phase 3: Multivariate Exploration & Correlations
*In this phase, I analyze how different factors (like Fare and Class) relate to survival using pairplots and heatmaps.*

### Creation of Family Size column

```python
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data.head(10)
```

### Creation of Age Group column

```python
condition = [
    (data['Age'] <= 12),
    (data['Age'] > 12) & (data['Age'] <= 19),
    (data['Age'] > 19) & (data['Age'] <= 59),
    (data['Age'] > 59)  
]

choice = ['Child','Teen','Adult','Senior']

data['Age Group'] = np.select(condition, choice, default='Unknown')
```

```python
data.head(10)
```

### Creation of Title column

```python
titles = ['Mr', 'Mrs', 'Miss', 'Master']
data['Title'] = data['Name'].str.extract('([A-Za-z]+)\.', expand=False)
# data['Title'] =data['Title'].apply(lambda x: x if x in titles else 'Others')
data.head(10)
```

### Creation of Isalone column

```python
data['Is Alone'] = data['FamilySize'].apply(lambda x: 'Alone' if x == 1 else 'Not Alone')
data.head(10)
```

```python
data.info()
```

## Phase 4: Statistical Insights
*In this final phase, I rank the features to determine which data has the strongest relationship with survival rates.*

### Histogram Plot for visualization of Overall Survival Percentage 

```python
ax = sns.histplot(data, x='Survived', stat='percent', shrink=0.8)
ax.set_xticks([0,1])
ax.set_xticklabels(['Died','Survived'])
plt.title('Overall Survival rate')
```

### Visual representation of Survival rate of each Gender

```python
ax = sns.countplot(data=data, x='Survived', hue='Sex', palette='tab10', width=0.8)
ax.set_xticks([0,1])
ax.set_xticklabels(['Died','Survived'])
plt.title('Survival Rate per gender')
```

### Visual representation of Survival of each Passenger class

```python
ax = sns.countplot(data=data, x='Survived', hue='Pclass', palette='tab10', width=0.8)
ax.set_xticks([0,1])
ax.set_xticklabels(['Died','Survived'])
plt.title('Survival Rate per class')
```

### Visual representation of Survival of each Age Group

```python
ax = sns.countplot(data=data, x='Survived', hue='Age Group', palette='tab10', width=0.8)
ax.set_xticks([0,1])
ax.set_xticklabels(['Died','Survived'])
plt.title('Survival Rate per Age Group')
```

### Visual representation of Survival per Family Size

```python
ax = sns.countplot(data=data, x='Survived', hue='FamilySize', palette='tab10', width=0.8)
ax.set_xticks([0,1])
ax.set_xticklabels(['Died','Survived'])
plt.title('Survival Rate by Family Size')
```

### Distribution of overall Age groups onboard

```python
sns.histplot(data=data, x='Age Group', shrink=0.5)
plt.title('Overall Age Group Distribution')
```

### Distribution of overall Fare 

```python
sns.histplot(data, x='Fare', kde=True,shrink=0.5)
plt.xlim(0,100)
plt.title('Overall fare Distribution')
```

### Distribution of overall Fare class wise and Detecting the outlier fare

```python
sns.boxplot(data=data, x='Pclass', y='Fare')
sns.set_style('whitegrid')
sns.set_palette('bright')
plt.title('Fare Distribution Classwise ( Identify outliers )')
```

### Distribution of overall Survival as compare to age

```python
sns.boxplot(data, x='Survived',y='Age')
```

### Relation Between Survival and differnt categories using heatmap

```python
corelation = data[['Survived','Age','Fare','FamilySize','Pclass']].corr()
sns.heatmap(data=corelation, annot=True, cmap='coolwarm')
```

### Relation Between Survival and differnt categories using pairplot

```python
sns.pairplot(data=data, vars=['Age','FamilySize','Fare'], hue='Survived')
```
